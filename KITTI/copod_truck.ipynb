```python
import os
import glob
import numpy as np
from pyod.models.copod import COPOD
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d

cars = np.load("output/gtruth_all.npy")

clf_name = 'COPOD'
clf = COPOD(contamination=0.01, n_jobs=3)
clf.fit(cars)

cars = []

i = 0
```

```python
predict_list = list()

incr = 100

if(i<1085):
    for j in range (i, i+incr):
        if(os.path.isdir("output/truck/points/")):
            file = glob.glob("output/truck/points/"+str(j)+"[_]*[_]*[.]npy")
            if(file):
                file = file[0]
                pts = np.load(file).reshape(-1, 3)

                predict = clf.predict(pts)
                outliers = predict[np.where(predict==1)]
                predict_list.append((outliers.size/predict.size)*100)
    
    if(i>100):
        predict_list = np.append(np.load("output/truck_predict.npy"), np.asarray(predict_list))

    i+=incr
    
    np.save("output/truck_predict.npy", predict_list)

    print("Batch #" + str(i/100) + " of 10.85 completed.")
    print(str(i))
```
